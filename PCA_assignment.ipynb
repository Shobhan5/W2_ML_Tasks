```python
train = pd.read_csv(r"C:\Users\Shobhan Sarkar\OneDrive\Desktop\ACM SIGCHI\W1-ML\bonus_data\train.csv")
```

```python
train.columns
```

```python
train.drop(['MoSold', 'YrSold'], axis=1, inplace=True)
```

```python
train = pd.get_dummies(train, drop_first=True)
train.head()
```

```python
numerical_df = train.select_dtypes(include=['number'])

# Check the resulting dataframe
print(numerical_df.columns)
```

```python
numerical_df.fillna(numerical_df.mean(), inplace=True)

# Check the resulting dataframe
print(numerical_df.head())
```

```python
pca = PCA(n_components=0.95)  # Keep 95% of variance
X_pca = pca.fit_transform(numerical_df)
```

```python
y = pd.get_dummies(train, drop_first=True)
y.head()
```

```python
numerical_df_test = y.select_dtypes(include=['number'])

# Check the resulting dataframe
print(numerical_df_test.columns)
```

```python
numerical_df_test.fillna(numerical_df_test.mean(), inplace=True)

# Check the resulting dataframe
print(numerical_df_test.head())
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, numerical_df_test, test_size=0.3, random_state=42)

# Initialize and train Linear Regression model
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Make predictions
y_pred = lin_reg.predict(X_test)

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Linear Regression Model:')
print(f'Mean Squared Error: {mse}')
print(f'R² Score: {r2}')
```
